```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv',sep=',')
```

1. Create a pie chart presenting the male/female proportion

```python
df.columns
```

```python
df.dropna(axis=0,subset=['sex'],inplace=True)
```

```python
df
```

```python
labels= ['Male','Female']
male = df.loc[df.sex=='male'].count()[0]
female = df.loc[df.sex=='female'].count()[0]

plt.style.use('ggplot')
weights= [male,female]

plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8)
plt.title('Gender Distribution (count%)')
plt.show()
```

```python
#df.info()
from sklearn import preprocessing
le= preprocessing.LabelEncoder()
values={'age':29,'fare':33}  #fill the null values by median
df.fillna(value=values,inplace=True)
df['sex']=le.fit_transform(df['sex'])
df.info()
```
