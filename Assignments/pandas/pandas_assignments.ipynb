5. In the RecentDelays column, the values have been entered into the
DataFrame as a list. We would like each first value in its own column, each 
second value in its own column, and so on. If there isn't an Nth value, the value
should be NaN.
Expand the Series of lists into a DataFrame named delays, rename the columns
delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df
with delays.


```python
df
```

```python
#write a function to expand the given column fro the given dataframe
```

```python
def expand_df(df,col):
    print("Obtain the highest possible delay column numbers")
    for i in range(len(df)):
        if(type(df.loc[i,col]) is list):
            n= len(df.loc[i,col])
            df.loc[i,'delayes_n']=n
    print('Expand the columns')
    col='RecentDelays'
    idx = df.sort_values(by='delayes_n',ascending=False).index[0]
    total_no_cols = df.loc[idx,'delayes_n'].astype(int)
    for i in range(len(df)):
        for j in range(total_no_cols):
            column = 'delay_' +str(j)
            try:
                df.loc[i,column] = df.loc[i,col][j]
            except:
                df.loc[i,column] = None
            
            
expand_df(df,'RecentDelays')
```
